```python
import pandas as pd
from scipy import stats
```

```python
#loading the dataset
ad_df = pd.read_csv('C:/Users/Admin/Downloads/ad_click_dataset.csv')
#removing nan values
ad_df = ad_df.dropna()
# converting categorical variables to numerical 
ad_df['gender'] = ad_df['gender'].map({'Male': 0, 'Female': 1})
ad_df['ad_position'] = ad_df['ad_position'].map({'Top': 0, 'Bottom': 1})
```

```python
ad_df
```

```python
#if ad position is top we calssify in group_a else group_b
group_a = ad_df[ad_df['ad_position'] == 0]  
group_b = ad_df[ad_df['ad_position'] == 1]
```

```python
#calculating the clicks
clicks_a = group_a['click'].sum()
total_a = len(group_a)
clicks_b = group_b['click'].sum()
total_b = len(group_b)
```

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = [clicks_a, clicks_b]
nobs = [total_a, total_b]
z_score, p_value = proportions_ztest(count, nobs, alternative='two-sided')
```

```python
train = pd.read_csv('train.csv')
test1 = pd.read_csv('test1.csv')
test2 = pd.read_csv('test2.csv')
```

```python
train
```

```python
# dropping nan values
train_no2 = train['NO2(GT)'].dropna()
test1_no2 = test1['NO2(GT)'].dropna()
test2_no2 = test2['NO2(GT)'].dropna()
```

```python
from scipy.stats import ks_2samp

# Test1 vs. Train
ks_stat1, pval1 = ks_2samp(train_no2, test1_no2)
print(f"Test1 vs. Train: KS Stat = {ks_stat1:.4f}, P-Value = {pval1:.4f}")

# Test2 vs. Train
ks_stat2, pval2 = ks_2samp(train_no2, test2_no2)
print(f"Test2 vs. Train: KS Stat = {ks_stat2:.4f}, P-Value = {pval2:.4f}")
```

```python
alpha = 0.05
shift_test1 = pval1 < alpha  # True if shift detected
shift_test2 = pval2 < alpha

print(f"Covariate Shift in Test1: {shift_test1}")
print(f"Covariate Shift in Test2: {shift_test2}")
```
